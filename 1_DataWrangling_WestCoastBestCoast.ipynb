## Step 3: Impute missing values for all columns

Define a missing value replacement method for each column (independently):

* **MODE**: Columns to replace NaN with mode/most frequent value:
    * bathroomcnt
    * bedroomcnt
    * roomcnt
    * fips
    * fullbathcnt
    * propertylandusetypeid
    * regionidcounty
    * unitcnt
    * numberofstories
    

* **ZEROS**: Columns to replace NaN with zeros (assuming missing data mean house does not have feature)
    * garagecarcnt (who owns a 25 car garage? https://www.latimes.com/business/realestate/hot-property/la-fi-hp-home-20180428-story.html) 
    * garagetotalsqft
    * lotsizesquarefeet (if missing, set to zero b/c average is very skewed, not normal distribution)
    
    
* **BINARY**: Columns to replace NaN with ZERO, all other values with ONE
    * hashottuborspa
    * poolcnt
    * taxdelinquencyflag
    
    
* **OTHER VALUE**: Columsn to replace NaN with specific categorical value
    * heatingorsystemtypeid: NaN -> 13 (meaning "None")
    * calculatedfinishedsquarefeet: NaN -> 1800 (average sq. ft. home size in Southern California)
    
* **NEW COLUMN**: 
    * age = 2015 - yearbuilt
    

```python
mode_fillcols = ['bathroomcnt','bedroomcnt','roomcnt','fips','fullbathcnt','propertylandusetypeid','regionidcounty','unitcnt','numberofstories']
zero_fillcols = ['garagecarcnt','garagetotalsqft','lotsizesquarefeet']
binary_replacecols = ['hashottuborspa','poolcnt','taxdelinquencyflag']
```

Using dataframe 4 `data4` (cleanest by this point)

```python
#MODE
for col in mode_fillcols:
    data4[col].fillna(data4[col].mode()[0], inplace=True)
```

```python
#ZEROS
for col in zero_fillcols:
    data4[col].fillna(0, inplace=True)
```

```python
#BINARY
for col in binary_replacecols:
#     data4[col].loc[~data4.isnull()] = 1
#     data4[col].loc[data4.isnull()] = 0
    data4[col] = data4[col].notnull().astype('int')
```

```python
#heatingorsystemtypeid
data4['heatingorsystemtypeid'].fillna(13, inplace=True)
#calculatedfinishedsquarefeet
data4['calculatedfinishedsquarefeet'].fillna(1800, inplace=True)
```

```python
#AGE COLUMN
#first, replace missing years with average year (48k missing years)
# mean_yearbuilt = data4['yearbuilt'].mean()
# = ~1964
data4['yearbuilt'].fillna(1964, inplace=True)
```

```python
data4['age'] = 2015-data4['yearbuilt']
```
